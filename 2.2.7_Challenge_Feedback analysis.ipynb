```python
import pandas as pd
import seaborn as sns
```

### Imdb

```python
imdb= pd.read_csv("/Users/OzlemCelik/Desktop/sentiment labelled sentences/imdb_labelled.txt", 
                  delimiter= '\t', header=None)
```

```python
imdb.columns = ['review', 'sentiment']

keywords = ['wasted', 'only', 
            'too', 'but', 
            'hated', 'disappointed', 
            'garbage', 'do not', 
            'bad', 'awful',
            'avoid', 'slow', 
            'insult', 'ridiculous',
            'lacks', 'predictable', 
            'doomed', 'crap', 
            'excellent', 'average',
            'BRILLIANCE', 'lazy',
            'mediocre', 'enjoyed',
            'glad','worst',
            'nerves', 'terrible',
            'horrible', 'best',
            'overrated', 'unfunny',
            'failed',
            'painful', 'no', 'stupid',
            'great', 'strong', 'nothing', 'aimless', 'predictable'
            ]

for key in keywords:
    # Note that we add spaces around the key so that we're getting the word,
    # not just pattern matching.
    imdb[str(key)] = imdb.review.str.contains(
        ' ' + str(key) + ' ',
        case=False
    )
    
```

```python
imdb['allcaps'] = imdb.review.str.isupper()
```

```python
sns.heatmap(imdb.corr())
```

```python
data = imdb[keywords + ['allcaps']]
target = imdb['sentiment']
```

```python
# Our data is binary / boolean, so we're importing the Bernoulli classifier.
from sklearn.naive_bayes import BernoulliNB

# Instantiate our model and store it in a new variable.
bnb = BernoulliNB()

# Fit our model to the data.
bnb.fit(data, target)

# Classify, storing the result in a new variable.
y_pred = bnb.predict(data)

# Display our results.
print("Number of mislabeled points out of a total {} points : {}".format(
    data.shape[0],
    (target != y_pred).sum()
))
```
